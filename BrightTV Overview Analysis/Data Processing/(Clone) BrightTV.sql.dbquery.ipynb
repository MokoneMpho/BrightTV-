```sql
select * from `workspace`.`default`.`viewership` limit 100;

select * from `workspace`.`default`.`userprofile` limit 100;


SELECT 
    P.UserID,
    P.Name,
    P.Surname,
    P.Province,
    v.Channel2,
    -- Gender bucket
    CASE 
        WHEN LOWER(P.Gender) IN ('male','m') THEN 'Male'
        WHEN LOWER(P.Gender) IN ('female','f') THEN 'Female'
        ELSE 'OtherOrUnknown'
    END AS Gender_group,

    -- Race bucket
    CASE 
        WHEN LOWER(P.Race) IN ('black','african') THEN 'Black'
        WHEN LOWER(P.Race) IN ('white') THEN 'White'
        WHEN LOWER(P.Race) IN ('coloured') THEN 'Coloured'
        WHEN LOWER(P.Race) IN ('indian','asian') THEN 'IndianOrAsian'
        ELSE 'OtherOrUnknown'
    END AS Race_group,


    -- Date & Time separation
    DATE(from_utc_timestamp(V.RecordDate2,'Africa/Johannesburg')) AS view_date,
    date_format(from_utc_timestamp(V.RecordDate2,'Africa/Johannesburg'),'HH:mm:ss') AS view_time,


CASE
        WHEN date_format(from_utc_timestamp(v.RecordDate2,'Africa/Johannesburg'), 'EEEE') IN ('Saturday','Sunday') THEN 'Weekend'
        WHEN date_format(from_utc_timestamp(v.RecordDate2,'Africa/Johannesburg'), 'EEEE') IN ('Monday','Tuesday','Wednesday','Thursday','Friday') THEN 'Weekday'
    END AS Day_classification,

CASE 
    WHEN date_format(from_utc_timestamp(V.RecordDate2,'Africa/Johannesburg'),'HH:mm:ss') 
         BETWEEN '05:00:00' AND '11:59:59'
    THEN 'Morning'

    WHEN date_format(from_utc_timestamp(V.RecordDate2,'Africa/Johannesburg'),'HH:mm:ss') 
         BETWEEN '12:00:00' AND '18:59:59'
    THEN 'Afternoon'

    WHEN date_format(from_utc_timestamp(V.RecordDate2,'Africa/Johannesburg'),'HH:mm:ss') 
         BETWEEN '19:00:00' AND '22:59:59'
    THEN 'Night'

    ELSE 'Hush of the Night'
END AS time_interval
FROM `workspace`.`default`.`viewership` V
LEFT JOIN `workspace`.`default`.`user_profiles` P
ON V.userid4 = P.UserID;
```
